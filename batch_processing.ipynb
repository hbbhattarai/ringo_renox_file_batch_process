Import Libraries

```python
import os
import subprocess
import glob
```

Define Paths

```python
base = r"D:\Master in Geoinformatics\Semester 2\GNSS Survey Mapping\Lab\Lab 2\ringo"

ringo = os.path.join(base, "ringo.exe")

data_path = os.path.join(base, "Data")
brdc_path = os.path.join(base, "BRDC")
output_path = os.path.join(base, "output")
```

Days

```python
days = ["1","2","3"]
```

Batch Script

```python
for day in days:
    print("Processing Day:", day)

    data_folder = os.path.join(data_path, day)
    brdc_folder = os.path.join(brdc_path, day)

    html_folder = os.path.join(output_path, day, "html")
    log_folder = os.path.join(output_path, day, "log")
    csv_folder = os.path.join(output_path, day, "rnxcsv")
    obs_folder = os.path.join(output_path, day, "obs")

    os.makedirs(html_folder, exist_ok=True)
    os.makedirs(log_folder, exist_ok=True)
    os.makedirs(csv_folder, exist_ok=True)
    os.makedirs(obs_folder, exist_ok=True)

    nav_files = glob.glob(os.path.join(brdc_folder, "*.gz"))

    if not nav_files:
        print("No BRDC file found for Day", day)
        continue

    nav_file = nav_files[0]
    print("BRDC File Used:", nav_file)

    obs_files = glob.glob(os.path.join(data_folder, "*.gz"))

    for obs in obs_files:

        station = os.path.basename(obs).split(".")[0]

        print("\nProcessing Station:", station)
        print("Observation File:", obs)
        print("Navigation File:", nav_file)

        log_file = os.path.join(log_folder, station + ".log")
        html_file = os.path.join(html_folder, station + ".html")
        csv_file = os.path.join(csv_folder, station + ".csv")
        obs_file = os.path.join(obs_folder, station + ".obs")

        # QC
        cmd = f'"{ringo}" qc "{obs}" "{nav_file}"'
        print("Running Command:", cmd)
        with open(log_file, "w") as f:
            subprocess.run(cmd, shell=True, stdout=f)

        # HTML
        cmd = f'"{ringo}" viewer --qcmode "{obs}" "{nav_file}"'
        print("Running Command:", cmd)
        with open(html_file, "w") as f:
            subprocess.run(cmd, shell=True, stdout=f)

        # OBS merge
        cmd = f'"{ringo}" merge --elmaskf 15 "{obs}"'
        print("Running Command:", cmd)
        with open(obs_file, "w") as f:
            subprocess.run(cmd, shell=True, stdout=f)

        # CSV
        cmd = f'"{ringo}" rnxcsv --qcmode "{obs}" "{nav_file}"'
        print("Running Command:", cmd)
        with open(csv_file, "w") as f:
            subprocess.run(cmd, shell=True, stdout=f)

    print("\nFinished Day", day)
```
